```python
import osmnx as ox
import matplotlib.pyplot as plt

# Location name
place = "Gila River Indian Community, Arizona, USA"

# Download boundary from OpenStreetMap
gdf = ox.geocode_to_gdf(place)

# Plot
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax, edgecolor="black", facecolor="lightblue")

plt.title("Gila River Indian Community")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
place = "Gila River Indian Community, Arizona, USA"

# Get boundary
boundary = ox.geocode_to_gdf(place)

# Get road network
graph = ox.graph_from_place(place, network_type="drive")

# Plot
fig, ax = plt.subplots(figsize=(12, 12))

boundary.plot(ax=ax, color="lightyellow", edgecolor="black")
ox.plot_graph(
    graph,
    ax=ax,
    node_size=0,
    edge_color="gray",
    show=False,
    close=False
)

plt.title("Road Map of Gila River Indian Community")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```

```python
import osmnx as ox
import geopandas as gpd
import matplotlib.pyplot as plt

place = "Gila River Indian Community, Arizona, USA"

# Get community boundary
boundary = ox.geocode_to_gdf(place)

# Get road network
graph = ox.graph_from_place(place, network_type="drive")

# Get water features
water_tags = {
    "natural": ["water"],
    "waterway": True
}

water = ox.features_from_place(place, water_tags)

# Create map
fig, ax = plt.subplots(figsize=(12, 12))

# Boundary
boundary.plot(
    ax=ax,
    facecolor="none",
    edgecolor="black",
    linewidth=2
)

# Water features
if not water.empty:
    water.plot(
        ax=ax,
        color="lightblue",
        alpha=0.7
    )

# Roads
ox.plot_graph(
    graph,
    ax=ax,
    node_size=0,
    edge_color="gray",
    edge_linewidth=0.5,
    show=False,
    close=False
)

plt.title("Gila River Indian Community\nRoads and Water Features")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.show()
```
